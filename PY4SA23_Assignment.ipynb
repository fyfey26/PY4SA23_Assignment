```python
from math import radians, cos, sin, asin, sqrt, atan2, degrees

def haversine (A,B):
    lat1=A[0]
    lon1=A[1]
    
    lat2=B[0]
    lon2=B[1]
    
    lat1,lon1,lat2,lon2=map(radians,[lat1,lon1,lat2,lon2])
    
    dlon=lon2-lon1
    dlat=lat2-lat1
    a=sin(dlat/2)**2+cos(lat1)*cos(lat2)*sin(dlon/2)**2
    c=2*atan2(sqrt(a),(sqrt(1-a)))
    R=6371
    return(R*c)
ans=haversine([2,1],[1,2])
print (ans)
    
                        
    
    
```
