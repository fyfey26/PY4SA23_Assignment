T3: A function was created to calculate the distance between 2 coordinates using the haversine formula. The coordinates ([2,1],[1,2]) were then used as inputs

```python
from math import radians, cos, sin, asin, sqrt, atan2, degrees

def haversine (A,B):
    lat1=A[0]
    lon1=A[1]
    
    lat2=B[0]
    lon2=B[1]
    
    lat1,lon1,lat2,lon2=map(radians,[lat1,lon1,lat2,lon2]) #converting values to radians
    
    dlon=lon2-lon1 #difference in longitude
    dlat=lat2-lat1 #difference in latitude
    a=sin(dlat/2)**2+cos(lat1)*cos(lat2)*sin(dlon/2)**2 #calculating formula for a
    c=2*atan2(sqrt(a),(sqrt(1-a))) #calculating formula for c
    R=6371 #Earth's radius
    return(R*c)
ans=haversine([2,1],[1,2]) # inputting pair of coordinates as two lists
print(ans)
    
    
    
                        
    
    
```
